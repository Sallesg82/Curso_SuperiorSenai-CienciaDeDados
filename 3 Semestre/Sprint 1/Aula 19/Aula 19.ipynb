```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Usando o Python

```python
pop = [20,30,40]

combinações = []

for x in pop:
    for y in pop:
        combinações.append([x, y])

print(combinações)
```

Usando Numpy

```python
comb = np.array(combinações)
comb


print(comb.var(axis=1, keepdims=True, ddof=1).mean())
print(comb.var(axis=1, keepdims=True, ddof=0).mean())
```
