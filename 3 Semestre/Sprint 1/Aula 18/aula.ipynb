```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
populacao = np.arange(1, 11)
populacao

def var(populacao):
  media = populacao.mean()
  soma = 0
  for x in populacao:
    soma = soma + (x-media)**2

  return soma / len(populacao)

def varam(amostra):
  media = amostra.mean()
  soma = 0
  for x in amostra:
    soma += (x - media)**2

  return soma / (len(amostra) - 1)

def dp(populacao):
  return var(populacao)**(1/2)

def dpam(amostra):
  return varam(amostra)**(1/2)

print(var(populacao), dp(populacao))

# amostra
print(dp(np.array([3,6,9])))
print(dpam(np.array([3,6,9])))
```
