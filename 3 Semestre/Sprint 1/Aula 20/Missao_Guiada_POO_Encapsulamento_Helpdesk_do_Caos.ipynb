## ✅ Missão 1 — Classe `Cliente`

### Regras
- `Cliente` tem `nome` e `telefone`
- Se `nome` vier vazio (ou só espaços), levante `ValueError`

### Você precisa implementar
- `__init__`
- `__repr__` (ou `__str__`) para ficar legível


```python
class Cliente:
    def __init__(self, nome: str, telefone: str):
        # TODO: validar nome (não pode vazio ou só espaços)
        if not nome.strip():
            raise ValueError("Opa, opa, opa. Você precisa colocar o nome, rapaz.")
        self.nome = nome
        self.telefone = telefone

    def __repr__(self) -> str:
        # TODO: retornar algo legível, por exemplo:
        # Cliente(nome='...', telefone='...')
        return f"Registrado cliente: '{self.nome}' do número: '{self.telefone}'"
```


## ✅ Missão 2 — Classe `Item` (o dispositivo do caos)

### Regras
- `Item` tem `nome` e `categoria` (ex.: `"Eletrodoméstico"`, `"Periférico"`, `"Místico"`)
- `nome` não pode ser vazio (nem só espaços)

### Você precisa implementar
- `__init__`
- `__repr__` (ou `__str__`)


```python
class Item:
    def __init__(self, nome: str, categoria: str):
        if not nome.strip():
            raise ValueError("Escreva o nome seu bobao")
        self.nome = nome
        self.categoria = categoria

    def __repr__(self) -> str:
        return f"O {self.nome} se encaixa como: '{self.categoria}'"
        
```

Aqui você junta tudo: `Cliente` + `Item` + as regras do sistema.

### Regras de negócio
- Chamado nasce com status **ABERTO**
- Diagnóstico começa como `None`

### Atributos mínimos
- `cliente: Cliente`
- `item: Item`
- `problema: str`
- `diagnostico: Optional[str]` (começa `None`)
- `status` (por enquanto pode ser `self.status = "ABERTO"` — depois vamos encapsular)

### Métodos
- `diagnosticar(texto)`
- `concluir()`

```python
class Chamado:
    def __init__(self, cliente: Cliente, item: Item, problema: str):
        self.cliente = cliente
        self.item = item
        self.problema = problema
        # TODO: diagnostico começa None
        self.diagnostico = None
        # TODO: status começa "ABERTO"
        self.status = "Aberto"

    def diagnosticar(self, texto: str) -> None:
        # TODO: validar texto (não pode vazio/só espaços)
        if not texto.strip():
            raise ValueError("O diagnostico não pode estar vazio ou espaçado")
        self.diagnostico = texto
        # TODO: salvar em self.diagnostico
        

    def concluir(self) -> None:
        # Regra: não conclui sem diagnóstico
        # TODO: se self.diagnostico for None -> raise ValueError("...")
        if self.diagnostico == None:
            raise ValueError("Não existe o ""Diagnostico")
        else:
            self.diagnostico = "CONCLUIDO"
        # TODO: senão -> mudar status para "CONCLUIDO"

    def resumo(self) -> str:
        # TODO: retornar uma string com status, cliente e item
        return f"OS | {self.status} | Cliente {self.cliente.nome} | Item: {self.item.nome} | Diagnostico: {self.diagnostico}"
        # Sugestão:
        # "OS | <STATUS> | Cliente: <nome> | Item: <nome> | Diagnóstico: <...>"
```


## ✅ Missão 6 — Encapsulamento (Parte 7)

Agora vem a regra: **ninguém muda status na mão**.

Você deve:
- trocar `status` por um atributo interno: `self._status`
- criar `@property def status(self): return self._status`
- **não criar setter**
- status só muda via métodos (`concluir`, etc.)

### Tarefa
1) Atualize a classe `Chamado`:
- onde era `self.status`, virar `self._status`
- crie a property `status` (somente leitura)
- ajuste `resumo()` para usar `self.status`

2) Teste obrigatório:
- tente fazer `ch1.status = "HACKEADO"`
- deve dar `AttributeError` e você captura com `try/except`


```python
# ✅ Refaça a classe Chamado AQUI (substituindo a anterior)
# Dica: você pode copiar/colar e ajustar.

class Chamado:
    def __init__(self, cliente: Cliente, item: Item, problema: str):
        # TODO: salvar cliente, item, problema
        self.cliente = cliente
        self.item = item
        self.problema = problema
        # TODO: diagnostico começa None
        self.diagnostico = None
        # TODO: status encapsulado começa "ABERTO" em self._status
        self._status = 'ABERTO'

    @property
    def status(self) -> str:
        return self._status

    def diagnosticar(self, texto: str) -> None:
        # TODO: validar e salvar diagnóstico
        self.diagnostico = texto


    def concluir(self) -> None:
        # TODO: raise se não tiver diagnóstico
        if self.diagnostico is None:
            raise ValueError("Diagnostico não pode estar vazio.")
        # TODO: mudar self._status para "CONCLUIDO"
        self._status = "Concluida"

    def resumo(self) -> str:
        # TODO: retornar string com self.status, cliente e item
        return f"Status: {self._status} | Cliente: {self.item} | Item: {self.item}"
```

### 🧪 Teste final (Missão 6)
Se passar aqui, você concluiu até Encapsulamento ✅

```python
# Recriar objetos (porque redefinimos a classe Chamado)
c1 = Cliente("Dona Cássia", "11 99999-0000")
i1 = Item("Geladeira que chora", "Eletrodoméstico")
ch1 = Chamado(c1, i1, "Toda vez que abre a porta ela faz drama.")

print(ch1.resumo())

# 1) tentar concluir sem diagnóstico
try:
    ch1.concluir()
except ValueError as e:
    print("Erro esperado ✅:", e)

# 2) diagnosticar e concluir
ch1.diagnosticar("Possível excesso de sentimentos no compressor.")
ch1.concluir()
print(ch1.resumo())

# 3) testar encapsulamento
try:
    ch1.status = "HACKEADO"
except AttributeError as e:
    print("Encapsulado ✅:", e)
```
