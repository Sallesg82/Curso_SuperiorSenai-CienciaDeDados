# 6) Ordem de Serviço (OS): juntando tudo

Agora vem a parte mais importante: **um objeto que protege regras**.

Uma OS precisa de:
- um cliente
- um dispositivo
- um status
- um diagnóstico (opcional no começo)

## 6.1 Regras (bem simples no começo)
1. A OS começa com status `ABERTA`
2. Só pode concluir se tiver diagnóstico

Vamos fazer a OS bem minimalista, e ir evoluindo.


```python
class Cliente:
    def __init__(self, nome: str, telefone: str):
        self.nome = nome
        self.telefone = telefone

    def resumo(self) -> str:
        return f"{self.nome} | tel: {self.telefone}"

    def __repr__(self):
        return f"Cliente(nome={self.nome!r}, telefone={self.telefone!r})"

class Dispositivo:
    def __init__(self, nome: str, marca: str, sintoma: str):
        self.nome = nome
        self.marca = marca
        self.sintoma = sintoma

    def testar(self) -> str:
        return (
            f"[TESTE] {self.nome} ({self.marca}) | "
            f"Sintoma: {self.sintoma}"
        )

    def __repr__(self) -> str:
        return (
            f"Dispositivo(nome={self.nome}, marca={self.marca}, sintoma={self.sintoma})"
        )

class OrdemDeServico:
  def __init__(self, cliente: Cliente, dispositivo: Dispositivo):
    self.cliente = cliente
    self.dispositivo = dispositivo
    self.status = 'ABERTA'
    self.diagnostico = None

  def registrar_diagnostico(self, texto: str) -> None:
    self.diagnostico = texto

  def concluir(self) -> None:
    #Regra: não conclui sem diagnóstico
    if self.diagnostico is None:
      raise ValueError('Não da para concluir sem diagnostico!')
    self.status = 'CONCLUIDA'

  def resumo(self) -> str:
    return (
        f"OS | {self.status} | Cliente: {self.cliente.nome} |"
        f"Dispositivo {self.dispositivo.nome}"
    )

ana = Cliente('Ana', '11 91111-1111')
carregador = Dispositivo('Carregador do Além', 'Genérica', 'só carrega em 37°C')
os1 = OrdemDeServico(ana, carregador)
os1.resumo()
```

```python
try:
    os1.concluir()
except ValueError as e:
    print('Erro esperado ✅:', e)
```

```python
os1.registrar_diagnostico('Solda fria + carregador com personalidade forte.')
os1.concluir()
os1.resumo()
```

Até aqui você já viu a essência da POO:
- O objeto guarda dados (atributos)
- O objeto faz ações (métodos)
- O objeto impõe regras

Ordem de serviço encapsulada

```python
class OrdemDeServico:
  def __init__(self, cliente: Cliente, dispositivo: Dispositivo):
    self.cliente = cliente
    self.dispositivo = dispositivo
    self._status = 'ABERTA'
    self.diagnostico = None

  @property
  def status(self) -> str:
    return self._status

  def registrar_diagnostico(self, texto: str) -> None:
    self.diagnostico = texto

  def concluir(self) -> None:
    #Regra: não conclui sem diagnóstico
    if self.diagnostico is None:
      raise ValueError('Não da para concluir sem diagnostico!')
    self._status = 'CONCLUIDA'

  def resumo(self) -> str:
    return (
        f"OS | {self.status} | Cliente: {self.cliente.nome} |"
        f"Dispositivo {self.dispositivo.nome}"
    )

os2 = OrdemDeServico(ana, carregador)
os2.resumo()
```

```python
try:
    os2.status = 'QUALQUER_COISA'
except AttributeError as e:
    print('Protegido ✅:', e)
```
