# 🧠 Parte 1 – O que é uma função?

Imagine que o sistema da loja precisa:
- mostrar mensagens
- calcular preços
- aplicar descontos
- verificar estoque

Reescrever código toda hora é um caos.

👉 **Funções existem para evitar o caos.**



## 📘 Definição simples

Uma **função** é:
> Um bloco de código com nome  
> que executa uma tarefa específica  
> sempre que for chamado


```python

def mensagem_boas_vindas():
    print("Bem-vindo à Loja de Descartáveis Seminovos!")

mensagem_boas_vindas()
```


# 🎒 Parte 2 – Funções com parâmetros

Parâmetros permitem que a função:
- receba informações
- trabalhe com dados diferentes


```python

def anunciar_produto(nome_produto):
    print("Hoje temos uma oferta imperdível de:", nome_produto)

anunciar_produto("Copo descartável levemente amassado")
anunciar_produto("Canudo ecológico veterano")
```


📌 Regra mental:
- **Parâmetro** → variável da função
- **Argumento** → valor enviado



# 🎁 Parte 3 – Funções com retorno

Algumas funções **calculam** algo e devolvem um valor.


```python

def calcular_preco(base, desgaste):
    preco_final = base - desgaste
    return preco_final

preco = calcular_preco(10, 3)
print("Preço final:", preco)
```


⚠️ Importante:
- `print()` mostra na tela
- `return` devolve um valor para ser usado depois



# 🧩 Parte 4 – Funções pequenas que ajudam

Sistemas reais são feitos de:
- várias funções pequenas
- cada uma com uma responsabilidade


```python

def verificar_qualidade(desgaste):
    if desgaste <= 2:
        return "Quase novo"
    elif desgaste <= 5:
        return "Seminovo honesto"
    else:
        return "Usado com orgulho"
```


# 🌀 Parte 5 – Funções chamando funções

Agora o sistema começa a ficar **profissional**.

Uma função pode usar outras funções.


```python

def gerar_relatorio_produto(nome, preco_base, desgaste):
    preco_final = calcular_preco(preco_base, desgaste)
    qualidade = verificar_qualidade(desgaste)

    print("Produto:", nome)
    print("Qualidade:", qualidade)
    print("Preço final: R$", preco_final)

gerar_relatorio_produto(
    "Guardanapo reutilizável",
    8,
    4
)
```


🧠 Observe:
- `gerar_relatorio_produto()` coordena o sistema
- As outras funções fazem o trabalho específico



# 🏪 Parte 6 – Mini Sistema da Loja

Agora vamos organizar tudo.


```python

def processar_produto(nome, preco_base, desgaste):
    mensagem_boas_vindas()
    gerar_relatorio_produto(nome, preco_base, desgaste)
    print("Obrigado por comprar lixo reutilizável conosco!")

processar_produto(
    "Copo descartável experiente",
    12,
    6
)
```


# 🧼 Boas práticas com funções

✔ Uma responsabilidade por função
