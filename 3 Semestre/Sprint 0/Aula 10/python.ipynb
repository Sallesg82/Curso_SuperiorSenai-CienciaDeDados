```python
# Diferenciando print de Return
def dobro_com_print(x): # Criando uma função chamada dobro_com_print que recebe x como parámetro
    print(f"Dobro {x * 2}")

def dobro_com_return(x):
    return x * 2
```

```python
dobro_com_print(30)
```

```python
dobro_com_return(30)
resultado = dobro_com_return(2) # Atribuindo o valor do return da função a variavel resultado.
print("Resultado aguardado:", resultado)
```
