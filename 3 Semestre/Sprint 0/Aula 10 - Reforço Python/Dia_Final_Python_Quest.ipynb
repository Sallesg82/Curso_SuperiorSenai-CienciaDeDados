# 1️⃣ FUNÇÕES — Reforço Definitivo (sem trauma)

## 📘 Modelo mental profissional
Pense em funções como **funcionários** da loja:

- Cada função tem **um cargo**
- Faz **uma tarefa específica**
- Recebe informações (parâmetros)
- Entrega um resultado (return)

### ✅ Boa prática essencial
> **Uma função deve fazer uma coisa só** (o máximo possível).

Isso deixa o sistema:
- mais legível
- mais fácil de testar
- menos bugado


## 1.1 `print` vs `return` (ponto crítico)
- `print()` mostra na tela (interface)
- `return` devolve um valor (lógica reutilizável)

| Situação | Use |
|---|---|
| Mostrar mensagem para usuário | `print()` |
| Devolver valor para outro código usar | `return` |


```python
def dobro_com_print(x):
    print("Dobro:", x * 2)

def dobro_com_return(x):
    return x * 2

dobro_com_print(5)          # mostra na tela
resultado = dobro_com_return(5)  # guarda valor
print("Resultado guardado:", resultado)
```

## 1.2 Funções auxiliares e composição (funções chamando funções)

A função orquestradora não deve calcular tudo.  
Ela deve **chamar funções auxiliares**.

Vamos começar com regras da loja:
- preço final depende de desgaste
- desgaste gera classificação nonsense


```python
def calcular_preco(preco_base: float, desgaste: int) -> float:
    # Regra simples: quanto mais desgaste, menor o preço
    return max(preco_base - desgaste, 0)

def classificar_produto(desgaste: int) -> str:
    if desgaste <= 2:
        return "Quase novo (só chorou um pouco)"
    elif desgaste <= 5:
        return "Seminovo honesto (sobreviveu)"
    else:
        return "Veterano de guerra (vende assim mesmo)"

def gerar_relatorio_produto(nome: str, preco_base: float, desgaste: int) -> dict:
    preco_final = calcular_preco(preco_base, desgaste)
    classe = classificar_produto(desgaste)
    return {
        "nome": nome,
        "preco_base": preco_base,
        "desgaste": desgaste,
        "classe": classe,
        "preco_final": preco_final,
    }

print(gerar_relatorio_produto("Copo descartável experiente", 12.0, 6))
```

## 1.3 Função orquestradora (organização do sistema)

A orquestradora coordena o fluxo:
- recebe dados
- chama auxiliares
- decide o que fazer
- retorna resultado pronto

> Ela é o “gerente” do sistema.


```python
def processar_cadastro_produto(nome: str, preco_base: float, desgaste: int) -> dict:
    # Gerente: usa funções auxiliares
    produto = gerar_relatorio_produto(nome, preco_base, desgaste)
    # Adiciona campo extra do "marketing"
    produto["slogan"] = f"{produto['classe']} — promoção imperdível (talvez)"
    return produto

produto_teste = processar_cadastro_produto("Guardanapo quase limpo", 8.0, 4)
produto_teste
```

# 2️⃣ TRATAMENTO DE ERROS — Domando o Vermelho da Tela

Erros comuns em sistemas de loja:
- usuário digita texto em vez de número (`ValueError`)
- produto inexistente (`KeyError`)
- arquivo não existe (`FileNotFoundError`)

## 📘 Modelo mental do `try/except`
- `try`: tente executar
- `except`: se der erro, não quebre o sistema — trate e continue


```python
def ler_preco_float(mensagem: str) -> float:
    while True:
        try:
            valor = float(input(mensagem))
            return valor
        except ValueError:
            print("😅 Opa! Isso não parece um número. Tente de novo (ex: 9.99).")

# Teste (digite coisas erradas de propósito):
# preco = ler_preco_float("Digite o preço do produto: ")
# print("Preço registrado:", preco)
```

Em vez de “ValueError”, prefira:

```python
def ler_inteiro_com_humor(mensagem: str) -> int:
    while True:
        try:
            return int(input(mensagem))
        except ValueError:
            print("🐍 Pythonildo diz: 'Isso aí não é inteiro, chefe.' Tenta de novo.")

# Teste:
# desgaste = ler_inteiro_com_humor("Digite o desgaste (0 a 10): ")
# print("Desgaste:", desgaste)
```

# 4️⃣ ESTRUTURAS DE DADOS APLICADAS — Listas e Dicionários de Verdade

## 📘 Regra mental
- **Lista**: coleção em ordem (vários itens)
- **Dicionário**: um item com atributos (chave → valor)

A loja vai usar:
- lista para guardar produtos cadastrados
- dicionário para representar cada produto


```python
# Nossa "base de dados" em memória (por enquanto)
produtos = []  # lista de dicionários

def adicionar_produto_na_lista(produto: dict) -> None:
    produtos.append(produto)

adicionar_produto_na_lista(processar_cadastro_produto("Canudo ecológico veterano", 5.0, 2))
adicionar_produto_na_lista(processar_cadastro_produto("Prato descartável reencarnado", 10.0, 7))

produtos
```

## 4.1 Manipulação real: update, loop e busca simples

Vamos:
- listar produtos
- atualizar preço de um produto (desconto)
- buscar por nome (simples)


```python
def listar_produtos():
    if not produtos:
        print("📦 Nenhum produto cadastrado ainda.")
        return
    for i, p in enumerate(produtos, start=1):
        print(f"{i}. {p['nome']} | classe: {p['classe']} | R$ {p['preco_final']:.2f}")

def aplicar_desconto_por_nome(nome: str, desconto: float):
    for p in produtos:
        if p["nome"].lower() == nome.lower():
            p["preco_final"] = max(p["preco_final"] - desconto, 0)
            p["slogan"] = p["slogan"] + " (com desconto!)"
            return True
    return False

listar_produtos()
print("Aplicando desconto...")
ok = aplicar_desconto_por_nome("Canudo ecológico veterano", 1.5)
print("Desconto aplicado?", ok)
listar_produtos()
```

## 5.1 Exportar `.txt` (log)
Bom para registrar eventos e reclamações.


```python
def salvar_log_txt(mensagem: str, caminho: str = "log_loja.txt"):
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    with open(caminho, "a", encoding="utf-8") as f:
        f.write(f"[{timestamp}] {mensagem}\n")

salvar_log_txt("Sistema iniciado.")
salvar_log_txt("Dois produtos cadastrados.")
print("✅ log_loja.txt atualizado!")
```
