# 🧪 PARTE 1 – O Sistema Acorda

## O que é programação?

Programar é **dar instruções claras** para um computador.  
O computador:
- Não adivinha
- Não interpreta intenções
- Executa exatamente o que está escrito

Python é uma linguagem que tenta ser **legível para humanos**, mas ainda exige precisão.


## 🔹 O comando `print()`

### Para que serve?
`print()` serve para **mostrar informações na tela**.

Ele não calcula.
Ele não salva.
Ele apenas **exibe**.


````python
print('Sistema Python inicializado com sucesso!')
````

## 🔹 Variáveis

### O que é uma variável?
Uma variável é um **espaço na memória** onde guardamos um valor.

Exemplo:
- Nome
- Idade
- Pontuação


### Regras importantes:
- O nome da variável **não pode ter espaços**
- Use letras minúsculas e `_`
- O `=` significa **atribuição**, não comparação


````python
nome = "Aventureiro"
nivel = 1

print("Bem-vindo,", nome)
print("Nível atual:", nivel)
````

## 🔹 Entrada de dados: `input()`

O `input()` permite que o usuário **digite algo**.

# 🧪 PARTE 2 – O Tribunal do Código

## Tomada de decisão

Programas tomam decisões usando **condições lógicas**.

A estrutura básica é:
```python
if condição:
    código
else:
    outro código
```


## 🔹 Operadores condicionais

- `>` maior que  
- `<` menor que  
- `>=` maior ou igual  
- `==` igual  
- `!=` diferente


````python
idade = int(input("Digite sua idade: "))

if idade >= 18:
    print("Acesso liberado")
else:
    print("Acesso negado")
````

## 🔹 IF encadeado (`elif`)

Usamos `elif` quando há **mais de uma possibilidade**.


````python
pontuacao = int(input("Digite sua pontuação: "))

if pontuacao >= 90:
    print("Classe S")
elif pontuacao >= 70:
    print("Classe A")
elif pontuacao >= 50:
    print("Classe B")
else:
    print("Classe C")
````

## ❌ Erro didático – sintaxe

O Python exige `:` no final da condição.


````python
if pontuacao > 50
    print("Pontuação válida")
````

# 🧪 PARTE 3 – O Loop Que Não Queria Parar

## O que é um laço de repetição?

Laços servem para **repetir código**.

Dois principais:
- `while`
- `for`


## 🔹 While

Repete enquanto a condição for verdadeira.


````python
contador = 1

while contador <= 5:
    print("Contador:", contador)
    contador += 1
````

## 🔹 For + range

Usado quando sabemos **quantas vezes** repetir.


````python
for i in range(1, 6):
    print("Iteração:", i)
````

## ⚠️ Loop infinito

Se a condição nunca muda, o loop **nunca termina**.


````python
while True:
    print("Socorro")
````

## 🎁 Missão bônus – Controle do tempo

Faça o sistema contar de trás para frente.


````python
for i in range(10, 0, -1):
    print(i)

print("Sistema estabilizado!")
````
